**Bank**


| Variable | Descripción |
|-------------|-------------|
| Edad (Age) | La edad de los clientes o prospectos que fueron contactados |
| Profesión (Job) | La profesión u ocupación de los clientes o prospectos |
| Estado civil (Marital) | El estado civil de los clientes o prospectos (soltero, casado, divorciado, etc.) |
| Nivel de educación (Education) |	El nivel de educación alcanzado por los clientes o prospectos. |
| Crédito por defecto (Default) |	Indica si los clientes tienen o han tenido algún crédito por defecto (sí, no) |
| Saldo promedio anual (Balance) |	El saldo promedio anual en la cuenta bancaria de los clientes |
| Préstamo de vivienda (Housing) | Indica si los clientes tienen un préstamo hipotecario (sí, no) |
| Préstamo personal (Loan) |	Indica si los clientes tienen un préstamo personal (sí, no) |
| Contacto (Contact) |	El método de contacto utilizado para comunicarse con los clientes o prospectos |
| Día de la semana del último contacto (Day) | El día de la semana en que se realizó el último contacto |
| Mes del último contacto (Month) |	El mes en que se realizó el último contacto |
| Duración del contacto (Duration) | La duración en segundos del último contacto realizado con los clientes o prospectos |
| Duración del último contacto (Campaign) | El número de veces que se ha contactado a los clientes o prospectos durante esta campaña |
| Días desde el último contacto (Pdays) |	El número de días transcurridos desde el último contacto realizado en una campaña anterior |
| Contactos realizados antes de campaña (Previous)  | El número de veces que se ha contactado a los clientes o prospectos antes de esta campaña |
| Resultado de la campaña anterior (Poutcome) | El resultado de la campaña de marketing anterior (si tuvo éxito, fracaso, etc.) |
| Suscripción a depósito a plazo fijo (Deposit) | Indica si los clientes se suscribieron o no a un depósito a plazo fijo (sí, no) |
